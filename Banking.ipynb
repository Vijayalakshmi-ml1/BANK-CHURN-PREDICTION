```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'C:/Users/Amrita/Downloads/Banking/banking_data.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
# Convert categorical variables to lowercase for consistency
df = df.apply(lambda x: x.astype(str).str.lower() if x.dtype == 'object' else x)

# Check for missing values
print("Missing values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=True, color='blue')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Age Among Clients")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='job', order=df['job'].value_counts().index, palette='viridis')
plt.xlabel("Count")
plt.ylabel("Job Type")
plt.title("Job Type Distribution Among Clients")
plt.show()
```

```python
# Marital Status Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='marital', palette='Set2')
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.title("Marital Status Distribution of Clients")
plt.show()
```

```python
# Education Level Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, y='education', order=df['education'].value_counts().index, palette='coolwarm')
plt.xlabel("Count")
plt.ylabel("Education Level")
plt.title("Education Level Distribution Among Clients")
plt.show()
```

```python
# Balance Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='default', palette='pastel')
plt.xlabel("Credit Default")
plt.ylabel("Count")
plt.title("Proportion of Clients with Credit in Default")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['balance'], bins=30, kde=True, color='purple')
plt.xlabel("Average Yearly Balance")
plt.ylabel("Count")
plt.title("Distribution of Average Yearly Balance Among Clients")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='housing', palette='muted')
plt.xlabel("Housing Loan")
plt.ylabel("Count")
plt.title("Number of Clients with Housing Loans")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='loan', palette='husl')
plt.xlabel("Personal Loan")
plt.ylabel("Count")
plt.title("Number of Clients with Personal Loans")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='contact', palette='dark')
plt.xlabel("Contact Type")
plt.ylabel("Count")
plt.title("Communication Types Used for Contacting Clients")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['day'], bins=31, kde=False, color='cyan')
plt.xlabel("Last Contact Day of the Month")
plt.ylabel("Count")
plt.title("Distribution of the Last Contact Day of the Month")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='month', order=['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'], palette='magma')
plt.xlabel("Month")
plt.ylabel("Count")
plt.title("Last Contact Month Variation Among Clients")
plt.show()

```

```python

plt.figure(figsize=(8, 5))
sns.histplot(df['duration'], bins=30, kde=True, color='green')
plt.xlabel("Last Contact Duration (seconds)")
plt.ylabel("Count")
plt.title("Distribution of Last Contact Duration")
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['campaign'], bins=20, kde=False, color='red')
plt.xlabel("Number of Contacts")
plt.ylabel("Count")
plt.title("Number of Contacts Performed During the Campaign")
plt.show()
```

```python
# Exclude '999' values as they indicate no previous contact
filtered_pdays = df[df['pdays'] != 999]['pdays']

plt.figure(figsize=(8, 5))
sns.histplot(filtered_pdays, bins=50, kde=True, color='brown')
plt.xlabel("Days Passed Since Last Contact")
plt.ylabel("Count")
plt.title("Corrected Distribution of Days Passed Since Last Contact")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['previous'], bins=20, kde=True, color='pink')
plt.xlabel("Number of Previous Contacts")
plt.ylabel("Count")
plt.title("Number of Contacts Performed Before Current Campaign")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='poutcome', palette='cool')
plt.xlabel("Previous Campaign Outcome")
plt.ylabel("Count")
plt.title("Outcomes of Previous Marketing Campaigns")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='y', palette='Set1')
plt.xlabel("Subscribed to Term Deposit")
plt.ylabel("Count")
plt.title("Subscription to Term Deposit")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Attributes")
plt.show()
```
